```python
import pandas as pd
from pathlib import Path

file_path = Path().resolve() # Path(__file__).resolve()
project_dir = file_path.parent
raw_data_path = project_dir / 'data' / 'raw'
processed_data_path = project_dir / 'data' / 'processed'
```

```python
xls = pd.ExcelFile(raw_data_path / 'Преаналитика и локализации.xlsx')
sheets_to_process = [sheet for sheet in xls.sheet_names if sheet != "Первый лист"]
unique_columns = set()

for sheet in sheets_to_process:
    df = pd.read_excel(xls, sheet_name=sheet, nrows=0) 
    unique_columns.update(df.columns.tolist())


sorted_columns = sorted(unique_columns)
for col in sorted_columns:
    print(col)
```
